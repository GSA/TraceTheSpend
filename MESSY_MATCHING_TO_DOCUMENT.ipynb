```python
contracts_file = [ f for f in glob.glob(f"data/usaspending/award_data/{agencydept}/*csv") if 'contracts' in f.lower()][0]
all_contracts =pd.read_csv(contracts_file)
#[col for col in list(all_contracts.columns) if 'agency' in col]
```

```python
program_cols = ['program_activities_funding_this_award',

 #'dod_acquisition_program_code',
 #'dod_acquisition_program_description',
 'small_business_competitiveness_demonstration_program',
 'program_acronym',
 'major_program',
 'the_ability_one_program',
 'c8a_program_participant']
office_agency_cols = [
#'federal_agency',
  #  'parent_award_agency_id',
 'parent_award_agency_name',
 #'awarding_agency_code',
 'awarding_agency_name',
# 'awarding_sub_agency_code',
 'awarding_sub_agency_name',
      #  'awarding_office_code',
 'awarding_office_name',
# 'funding_office_code',
 'funding_office_name',
# 'funding_agency_code',
 'funding_agency_name',
# 'funding_sub_agency_code',
 'funding_sub_agency_name',
 
]
cols_of_interest = [
 'award_id_piid',
 'usaspending_permalink',]
```

```python
# sorted(all_contracts[all_contracts["major_program"].notna()]["major_program"].unique())
all_contracts['program_actvity_name_code'] = all_contracts['program_activities_funding_this_award'].map(lambda x: x.split(";"), na_action="ignore")
```

```python
all_contracts.to_dict("records")[0]
```

```python

contracts2fpi = (
    dedup.merge(
        all_contracts.explode('program_actvity_name_code'),
        right_on='program_actvity_name_code', left_on='program_actvity_name_code',
        how="inner")[list(dedup.columns) + cols_of_interest + office_agency_cols ].drop_duplicates()
)#.to_csv("data/outputs/contract-award-matches-dept-of-justice-DOJ.csv", index=False)
```

```python
print(contracts2fpi.shape[0])
contracts2fpi[["fpi_al_number",] + cols_of_interest].drop_duplicates()
print(contracts2fpi["award_id_piid"].nunique())
print(contracts2fpi["usaspending_permalink"].nunique())
#contracts2fpi.sort_values("fpi_al_number")

```

```python

merged_table_fpi_program2contracts = fpi_all_program_dept_interest[["al_number", "title", "popular_name", "agency", "sub-agency", ]].drop_duplicates().merge(
    contracts2fpi[["fpi_al_number", "award_id_piid", "usaspending_permalink"]].drop_duplicates(),
    how="right", right_on = "fpi_al_number", left_on="al_number"
)
merged_table_fpi_program2contracts.to_excel(f"~/Downloads/example-fpi-to-contracts-{agencydept}.xlsx", index=False)
merged_table_fpi_program2contracts
```
