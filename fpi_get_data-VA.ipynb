```python
import re

def get_match(text, rex):
    if isinstance(rex, (list, tuple, set)):
        # Escape each program name individually
        escaped_rex = [re.escape(name) for name in rex]
        rex = "(" + "|".join(escaped_rex) + ")"
    result = re.findall(rex, text)
    return result

import datetime
print(datetime.datetime.now())

df_list = []
for pa in unique_program_activity_names_list:
    matches = get_match(text=pa.lower().strip(), rex=[a.lower().strip() for a in set_program_names])
    df_list.append({"program_activity_name": pa, "matches": matches, "#matches": len(matches)})
print(datetime.datetime.now())

    
    
                                                              
```

```python
match_df = pd.DataFrame(df_list)

program_activity_name_list = match_df[match_df["#matches"]>=1]["program_activity_name"].tolist()

usaspending_account_df_matches = usaspending_accountbreakdown[usaspending_accountbreakdown["program_activity_name"].isin(program_activity_name_list)]

# #.drop_duplicates()
```

```python
program_name_usa = usaspending_account_df_matches["program_activity_name"].unique().tolist()
print(len(program_name_usa))
list(program_name_usa)

final_ = match_df[match_df["#matches"]!=0]
final_program_names = final_.explode("matches")["matches"].dropna().unique()
print(f"{len(final_program_names)} fpi program names have matches to usaspending accounts")

FINAL_MATCH_DF = fpi_all_program[(fpi_all_program["agency"]==AGENCY_IDENTIFIER_NAME) & fpi_all_program["title"].map(lambda x: True if str(x).lower().strip() in final_program_names else False)]
```

```python
final_["fpi_program_name"] = final_["matches"].map(lambda x: x[0])
usaspendingprogramactivityname2fpiprogramname = final_.set_index(["program_activity_name"])["fpi_program_name"].to_dict()
usaspendingprogramactivityname2fpiprogramname
```

```python
FINAL_MATCH_DF["title_lower"] = FINAL_MATCH_DF["title"].map(lambda x: str(x).lower().strip())
fpiprogramname2alnumber = FINAL_MATCH_DF.set_index("title_lower")["al_number"].to_dict()
```
